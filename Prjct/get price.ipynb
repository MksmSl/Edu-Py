```python
import requests
```

```python
resp = requests.get('https://store.tildacdn.com/api/getproductslist/?storepartuid=488842231311&recid=345443903&c=1691240585349&getparts=true&getoptions=true&slice=1&size=300')
```

```python
print(resp.status_code)
print(type(resp.json()))
print(resp.json()["products"])
s = "Give Love Back"
for i in resp.json()["products"]:
    if s == i['title']:
        print(i['price'])
        print(i)
        
```
