```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")

titanic.head()
```

```python
fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(16, 5)

# sns.displot(titanic, x='age', kind='kde')

# Alternative for sns.displot, which cannot be used with custom axes, since it's figure-level
sns.histplot(titanic['age'], bins=10, ax=axs[0])
sns.kdeplot(titanic['age'], fill=True, bw_adjust=.5, cut=0, ax=axs[1])

# Empirical cumulative distributions
_ = sns.ecdfplot(titanic['age'], ax=axs[2])
```

```python

_ = sns.regplot(x="fare", y="age", data=titanic)
```

```python
_ = sns.catplot(y="embarked", x="age",  data=titanic, kind='strip')
```

```python
_ = sns.catplot(y="fare", x="pclass", kind="swarm", data=titanic)
```

```python
_ = sns.catplot(y="fare", x="pclass",  data=titanic, kind='strip')
```

```python
_ = sns.catplot(x="pclass", y="age", hue="sex", kind="box", data=titanic)
```

```python
_ = sns.catplot(x="fare", y="sex", hue="survived", kind="violin", orient="h", data=titanic)
```
